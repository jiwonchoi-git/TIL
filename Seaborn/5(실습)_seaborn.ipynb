## 1. 공유 자전거 데이터 더 살펴보기 I
> - 실습 내용
>
> 지난 영상에서 사용한 공유 자전거 데이터를 가지고 계속 연습을 해 봅시다. 이번에는 분기별로 정기권을 이용하는 사람들의 평균 대여 건수를 막대 그래프로 그려 볼 건데요. 아래와 같이 영업일과 영업일이 아닌 날을 한눈에 비교할 수 있게 코드를 작성해 주시면 됩니다.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bike_df = pd.read_csv('../data-2/bike.csv')
sns.set_theme(rc={'figure.figsize': (10, 5)}, style='white') # 그래프의 전역 설정
```

```python
bike_df.head(3)
```

```python
sns.barplot(data=bike_df, x='quarter', y='registered', hue='workingday', ci=None)
plt.show()
```

## 6. 보험료 데이터 분석하기
> - 실습 내용
>
> 보험료 인상에 영향을 주는 요인은 무엇인지 확인해 봅시다. 보험사 고객들의 건강 정보와 보험료 금액을 모아 놓은 insurance_df를 가지고 아래와 같은 히트맵을 그려 보세요.

```python
insurance_df = pd.read_csv('../data-2/insurance_premium.csv')
sns.set_theme(rc={'figure.figsize': (8, 6)}, style='white')
insurance_df.head(3)
```

```python
numeric_df = insurance_df.select_dtypes(include=[int, float])
sns.heatmap(numeric_df.corr(), annot=True) # annot 파라미터:  상관계수값까지 레이블링
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 특정 열의 상관 관계만 추출
column_corr = numeric_df.corr()[["premium_price"]].drop("premium_price", axis=0)

# 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(column_corr, annot=True, cmap="coolwarm")
plt.show()
```
