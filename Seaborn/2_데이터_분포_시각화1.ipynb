# 데이터 분포 시각화1

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/bike.csv') # 자전거의 이용 데이터
df
```

### stripplot vs scatterplot

- 전제: 데이터의 수가 `수십~수백 정도의 작은` 데이터셋일 떄, 사용
    - 점을 찍어 본다는 것 자체가 대용량의 데이터일 경우에는 표현력이 떨어진다는 뜻.

#### 1. stripplot
- **설명**: **`범주형 데이터`** 를 시각화할 때 사용. 각 데이터 포인트를 점 형태로 범주형 값에 맞게 흩뿌려 표현.
- **특징**: 범주형 변수와 연속형 변수 간의 관계를 점으로 표시하며, `jitter` 옵션을 사용해 겹치는 데이터를 분리 가능.

#### 2. scatterplot
- **설명**: **`두 개의 연속형 변수 간`**의 관계를 시각화할 때 사용. 2D 평면에서 각 데이터 포인트의 위치가 x, y축 값에 따라 결정됨.
- **특징**: 변수 간의 **상관관계** 를 시각화하는 데 유용하며, 두 연속형 변수 간의 패턴을 탐색할 때 사용.

#### 차이점
- **데이터 유형**: stripplot은 `범주형과 연속형` 데이터를, scatterplot은 `연속형 데이터 간`의 관계를 시각화.
- **축의 역할**: stripplot은 `x축에 범주형` 변수를, scatterplot은 x축과 y축 `모두에 연속형` 변수를 사용.
- **목적**: stripplot은 `범주 내 데이터 분포`를 시각화, scatterplot은 `두 변수 간의 상관관계`를 확인하는 데 사용.

```python
sns.stripplot(data=df, x='month', y='total', hue='workingday', jitter=True)
plt.show()
# jitter, 데이터가 너무 겹쳐보일 땐, 이를 피하기 위해 약간의 노이즈를 추가합니다.
```

- `swarmplot` : stripplot에 겹치는 점들을 옆으로 지렁이처럼 나열한 차이

```python
sns.swarmplot(data=df, x='month', y='total', hue='workingday')
plt.show()
```

> - swarmplot을 사용해서 중첩된 점들을 가로로 쭉 나열하는 식으로 대체를 했다.
>    - 그럼에도 모든 개수의 점들을 나타낼 수가 없는 문제가 있다.
>       1. 그렇기에, 대안으로 stripplot에 jitter를 추가하거나, 
>       2. violinplot을 사용해보라고 제안 -> 다음에 더 자세히 다룰 예정

```python
# 마커 크기를 줄이는 방법
### 분포가 더 잘 나타난다!
sns.violinplot(data=df, x="month", y="total")
plt.show()
```
