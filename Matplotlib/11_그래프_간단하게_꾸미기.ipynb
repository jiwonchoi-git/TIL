# 그래프 간단한게 꾸미기
- Matplotlib 그래프 꾸미기

## 1. 기본 데이터 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 키와 몸무게 데이터
height_array = np.array([
	165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
	165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
	172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
	62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
	57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
	67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

```python
height_array
```

```python
weight_array
```

## 2. 산점도 기본 출력

```python
 # x축 키, y축 몸무게
plt.scatter(height_array, weight_array)
plt.show()
```

## 3. 제목 붙이기
- `plt.title(제목 이름)`

```python
plt.scatter(height_array, weight_array)
plt.title("Height and Weight")
plt.show()
```

## 4. 축 이름 추가
- `plt.xlabel(x축 이름)`
- `plt.ylabel(y축 이름)`

```python
plt.scatter(height_array, weight_array)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

## 5. 점 색 바꾸기
- 옵션 `c` (color map)

- 점의 색상을 빨간색으로 변경

```python
plt.scatter(height_array, weight_array, c='Red')
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
```

- 점의 색상을 초록색으로 변경

```python
plt.scatter(height_array, weight_array, c='Green')
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
```

- Hex 코드로 색상 지정
    - 구글링 또는 GPT 사용해서, 찾기

```python
plt.scatter(height_array, weight_array, c='#6500C3')
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
```

## 6. 점 모양 바꾸기
- 옵션 `marker`

- `+` 모양으로 변경

```python
plt.scatter(height_array, weight_array, c='#6500C3', marker='+')
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
```

- 네모 모양으로 변경 : `s(square)`

```python
plt.scatter(height_array, weight_array, c='#6500C3', marker='s')
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
```

*  다양한 마커 옵션은 [Matplotlib Markers](https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers) 에서 확인할 수 있습니다.
