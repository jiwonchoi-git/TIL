# 그래프에 한글로 된 텍스트 넣기
-  Matplotlib에서 그래프의 제목이나 축 레이블에 `한글을 사용하면 글자가 깨지는 문제`가 있습니다.
- 이를 어떻게 해결하면 좋을까요?

## 1. 문제 설명
- 기본적으로 Matplotlib에서 사용하는 폰트는 **영어용 폰트**로, 한글을 제대로 표시하지 못하는 경우가 발생합니다.
- `그래프에 한글을 넣으면 글자가 깨져 보이거나 오류가 발생`할 수 있습니다.

## 2. 해결 방법: 폰트 설정 변경
- Matplotlib에서 한글을 제대로 표현하려면 `폰트를 한글을 지원하는 폰트로 변경`해야 합니다.
- 이를 위해 `plt.rc(속성인자, family=폰트의 계열)` 함수를 사용하여 폰트를 설정합니다.

## 3. Windows에서 폰트 변경
- Windows 사용자는 `맑은 고딕(Malgun Gothic)` 폰트를 사용할 수 있습니다.
- 이 코드를 사용하여 폰트를 변경한 후 그래프를 그립니다.

```python
import numpy as np
import matplotlib.pyplot as plt

height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
```

## 4. Mac에서 폰트 변경
- Mac 사용자는 기본적으로 설치된 `‘AppleGothic’` 폰트를 사용할 수 있습니다.

```python
plt.rc('font', family='AppleGothic')  # 폰트 설정
plt.scatter(height_array, weight_array)  # 데이터 그래프화
plt.title('키와 몸무게')  # 그래프 제목
plt.xlabel('키(cm)')  # x축 레이블
plt.ylabel('몸무게(kg)')  # y축 레이블
plt.show()  # 그래프 출력
```

## 5. 정리
- `폰트 설정을 변경`하면 한글이 깨지지 않고 정상적으로 표시됩니다.
