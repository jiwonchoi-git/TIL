## 1. 기본 DataFrame 준비
- 예시로 분기별 매출과 비용이 들어간 DataFrame을 생성합니다.

```python
import pandas as pd
```

```python
sales_df = pd.DataFrame({
    'quarter': ['1Q', '2Q', '3Q', '4Q'],
    'revenue': [1360, 2650, 2070, 4150],
    'cost': [1240, 1970, 1750, 2760]
})
```

```python
sales_df
```

## 2. matplotlib.pyplot을 사용한 기본 선 그래프
- matplotlib.pyplot을 사용해 x축에 분기(quarter), y축에 매출(revenue)을 넣어 선 그래프를 그립니다.

```python
import matplotlib.pyplot as plt
```
