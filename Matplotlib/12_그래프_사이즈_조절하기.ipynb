# 그래프 사이즈 조절하기
- Matplotlib을 사용한 그래프 크기 조절

````python
import numpy as np
import matplotlib.pyplot as plt

height_array = np.array([
    165, 164, 155, 151, 157, 162, 155, 157, 165, 162,
    165, 167, 167, 183, 180, 184, 177, 178, 175, 181,
    172, 173, 169, 172, 177, 178, 185, 186, 190, 187
])

weight_array = np.array([
    62, 59, 57, 55, 60, 58, 51, 56, 68, 64,
    57, 58, 64, 79, 73, 76, 61, 65, 83, 80,
    67, 82, 88, 62, 61, 79, 81, 68, 83, 80
])
````

## 1. 기본 그래프 사이즈
- Matplotlib 기본 단위 인치(inch) : 2.54cm
- 기본 값 : `가로 X 세로 = 6 X 4 (inch)`

````python
plt.scatter(height_array, weight_array)
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
````

## 2. 그래프 사이즈 조절하기
- `plt.figure(figsize=(가로 길이, 세로 길이))` : inch 단위
- 이 방법은 각각의 그래프에 **일시적**으로 적용된다.

````python
plt.figure(figsize=(10,4))
plt.scatter(height_array, weight_array)
plt.title("Height and Weight") # 제목 추가
plt.xlabel("Height (cm)") # x축 이름 추가
plt.ylabel("Weight (kg)") # y축 이름 추가
plt.show()
````

- 전체적으로 그래프 크기를 통일
    - `plt.rcParams` 속성에 접근해서 변경
    - rcParams는 **runtime configuration parameters**의 준말
        - Matplotlib의 런타임 설정 관리
        1. 그래프의 크기
        2. 폰트
        3. 색상
        4. 스타일
        - 등 다양한 설정들
>    ex) 
>    ```python
>    plt.rcParams['figure.figsize'] = (5,5)
>    ```

````python
plt.rcParams['figure.figsize'] = (5,5)
plt.scatter(height_array, weight_array)
plt.title("Height and Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
````
